```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
```

ARCHIVO DE VALENCIA

```python
# Ruta de la carpeta donde están los CSVs
carpeta = "C:/Users/Usuario/Desktop/UFV/4-CUARTO/TFG/DATOS/4. FINALES (csv)/LonjaValencia"

# Listar todos los archivos CSV en la carpeta
archivos = [f for f in os.listdir(carpeta) if f.endswith('.csv')]

print(archivos)
```

```python
# Verificar estacionariedad y analizar componentes para las columnas 'Máximo' y 'Mínimo'
for archivo in archivos:
    ruta_archivo = os.path.join(carpeta, archivo)
    
    # Cargar datos correctamente
    df = pd.read_csv(ruta_archivo, sep=";", parse_dates=['Fecha'], index_col='Fecha', dayfirst=True)

    for columna_valor in ['Máximo', 'Mínimo']:
        if columna_valor not in df.columns:
            print(f"La columna {columna_valor} no está en {archivo}, se omite.")
            continue
        
        # Verificar si la columna solo tiene NaN
        if df[columna_valor].isna().all():
            print(f"La columna '{columna_valor}' en {archivo} tiene solo NaN. Se omite.")
            continue
        
        # Interpolar valores NaN
        df[columna_valor] = df[columna_valor].interpolate(method='linear')

        # Verificar si la serie es constante
        if df[columna_valor].nunique() <= 1:
            print(f"La columna {columna_valor} en el archivo {archivo} es constante. Se omite la prueba.")
            continue
        
        # Asegurar que no haya NaN antes de la prueba ADF
        serie = df[columna_valor].dropna()
        if len(serie) == 0:
            print(f"La columna {columna_valor} en el archivo {archivo} no tiene datos suficientes tras limpieza. Se omite.")
            continue
        
        # Prueba de estacionariedad ADF
        print(f"\nPrueba de estacionariedad para '{columna_valor}' en {archivo}")
        resultado = adfuller(serie)
        
        print("Resultados de la Prueba Dickey-Fuller Aumentada:")
        print(f"Estadístico ADF: {resultado[0]}")
        print(f"p-valor: {resultado[1]}")
        print("Valores Críticos:")
        for clave, valor in resultado[4].items():
            print(f"   {clave}: {valor}")
        if resultado[1] <= 0.05:
            print(f"La columna '{columna_valor}' es estacionaria (rechazamos H0).")
        else:
            print(f"La columna '{columna_valor}' no es estacionaria (no rechazamos H0).")
        
        # Limpiar datos para descomposición
        serie_limpia = serie.dropna()
        if len(serie_limpia) < 12:  # Evita error si hay pocos datos
            print(f"La serie en '{archivo}' no tiene suficientes datos tras limpieza. Se omite el análisis.")
            continue
        
        # Análisis de las 4 componentes de la serie temporal
        decomposition = seasonal_decompose(serie_limpia, model='additive', period=12)
        
        # Extraer componentes
        tendencia = decomposition.trend
        estacionalidad = decomposition.seasonal
        residuo = decomposition.resid
        
        # Graficar las componentes
        fig, axes = plt.subplots(1, 4, figsize=(16, 4))
        fig.suptitle(f"Descomposición de '{columna_valor}' en {archivo}", fontsize=14)
        
        axes[0].plot(serie_limpia, label='Serie Original', color='black')
        axes[0].set_title('Serie Original')
        axes[0].legend(loc='upper left')
        
        axes[1].plot(tendencia, label='Tendencia', color='blue')
        axes[1].set_title('Tendencia')
        axes[1].legend(loc='upper left')
        
        axes[2].plot(estacionalidad, label='Estacionalidad', color='green')
        axes[2].set_title('Estacionalidad')
        axes[2].legend(loc='upper left')
        
        axes[3].plot(residuo, label='Ruido (Irregularidad)', color='red')
        axes[3].set_title('Ruido')
        axes[3].legend(loc='upper left')
        
        plt.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()
        
        print(f"Análisis completado para '{columna_valor}' en {archivo}\n")
```

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
import pandas as pd

# Especificar la ruta del archivo CSV
ruta_archivo = "C:/Users/Usuario/Desktop/UFV/4-CUARTO/TFG/DATOS/4. FINALES (csv)/datosAemet.csv"  # Archivo actualizado

# Cargar datos
df = pd.read_csv(ruta_archivo, sep=';', parse_dates=['Fecha'], index_col='Fecha', dayfirst=True)
```

```python
for columna_valor in ['TemperaturaMedia', 'Precipitación00-24h(mm)']:
    if columna_valor not in df.columns:
        print(f"La columna {columna_valor} no está en el archivo, se omite.")
        continue
    
    # Convertir a numérico manejando errores
    df[columna_valor] = pd.to_numeric(df[columna_valor], errors='coerce')
    
    # Interpolar valores NaN
    df[columna_valor] = df[columna_valor].interpolate(method='linear')
    
    # Verificar si la serie es constante
    if df[columna_valor].nunique() <= 1:
        print(f"La columna {columna_valor} es constante o no tiene suficiente variación. Se omite la prueba.")
        continue
    
    # Asegurarse de eliminar NaN restantes después de la interpolación
    serie = df[columna_valor].dropna()
    
    if len(serie) == 0:
        print(f"La columna {columna_valor} no tiene suficientes datos para la prueba. Se omite.")
        continue
    
    # Prueba de estacionariedad
    print(f"\nPrueba de estacionariedad para la columna '{columna_valor}'")
    resultado = adfuller(serie)
    
    print("Resultados de la Prueba Dickey-Fuller Aumentada:")
    print(f"Estadístico ADF: {resultado[0]}")
    print(f"p-valor: {resultado[1]}")
    print("Valores Críticos:")
    for clave, valor in resultado[4].items():
        print(f"   {clave}: {valor}")
    if resultado[1] <= 0.05:
        print(f"La columna '{columna_valor}' es estacionaria (rechazamos H0).")
    else:
        print(f"La columna '{columna_valor}' no es estacionaria (no rechazamos H0).")
    
    # Asegurarse de eliminar valores NaN restantes para la descomposición
    serie_limpia = serie.dropna()
    
    # Análisis de las 4 componentes de la serie temporal
    decomposition = seasonal_decompose(serie_limpia, model='additive', period=12)  # Ajusta 'period' según los datos
    
    # Extraer componentes
    tendencia = decomposition.trend
    estacionalidad = decomposition.seasonal
    residuo = decomposition.resid
    
    # Graficar las componentes
    fig, axes = plt.subplots(1, 4, figsize=(16, 4))
    fig.suptitle(f"Descomposición de '{columna_valor}'", fontsize=14)
    
    axes[0].plot(serie_limpia, label='Serie Original', color='black')
    axes[0].set_title('Serie Original')
    axes[0].legend(loc='upper left')
    
    axes[1].plot(tendencia, label='Tendencia', color='blue')
    axes[1].set_title('Tendencia')
    axes[1].legend(loc='upper left')
    
    axes[2].plot(estacionalidad, label='Estacionalidad', color='green')
    axes[2].set_title('Estacionalidad')
    axes[2].legend(loc='upper left')
    
    axes[3].plot(residuo, label='Ruido (Irregularidad)', color='red')
    axes[3].set_title('Ruido')
    axes[3].legend(loc='upper left')
    
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
    
    print(f"Análisis completado para la columna '{columna_valor}'\n")
```
